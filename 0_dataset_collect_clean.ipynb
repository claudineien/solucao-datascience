### COLETAR VÍDEOS DO YOUTUBE

### Bibliotecas com métodos, formulas, rotinas para agilizar desenvolvimento

```python
import pandas as pd
import numpy as np
import re
import time
import youtube_dl
```

### Definir palavras chave para pesquisa e quantidade de vídeos a localizar

```python
# Definir palavras chave para buscar os vídeos no youtube
keyword = ["machine+learning","data+science","kaggle","artificial intelligence"]
# Quantidade de vídeos a baixar, limitado a 20 vídeos por consulta
nqtdvid = '30'
```

### Coletar os vídeos do youtube

```python
# Pega os vídeos do youtube e grava na variavel resultado
resultado = []
# Pegar dados somente dos vídeos que estão na variável keyword
for qry in keyword:
    # Faz o tratamento de exções de erro
    try:    
        # Instanciar objeto YoutubeDL() com alguns argumentos configurados
        # ignoreerrors:True serve para Continuar o download ainda que exista os erros:
        #  1 DownloadError: ERROR: This video is unavailable
        ydl = youtube_dl.YoutubeDL({"ignoreerros":True})
        # Pegar os NNN últimos vídeos
        dados = ydl.extract_info("ytsearchdate"+nqtdvid+":{}".format(qry),download=False)
        # Pegar da estrutura entries os dados diretamente relacionados aos vídeos
        for entry in dados['entries']:
            # Considerar apenas os vídeos diferentes de None
            if entry is not None:
                # Gravar o conteudo da keyword desejada na nova coluna 'query'
                entry['query'] = qry
        # Gravar todo o conteudo do youtube na variavel resultado
        resultado += dados['entries']
    except:
            print("Could not download the video in course")
```

```python
# Mostrar o conteudo da estrutura do youtube, para identificarmos as colunas que precisaremos
# Importante : esta rotina pode demorar muito, depende da quantidade de vídeos informados em nqtdvid = '30'
len(resultado)
```

```python
# Armazenar os dados coletados sobre os vídeos em um dataframa
df = pd.DataFrame(resultado)
```

### Tratar alguns dados e incluir como novas colunas no data frame para melhor uso no modelo de predição

```python
# Incluir colunas com
# converter de tipo objeto para data a coluna existente upload_date, para conseguirmos utilizar como data
df['upload_date'] = pd.to_datetime(df['upload_date'])
# quantidade de dias que o video esta disponibilizado no youtube
df['tempo_desde_pub'] = (pd.to_datetime('2020-08-23')-df['upload_date'])/np.timedelta64(1,'D')
```

```python
# Configurar argumento com a quantidade total de colunas do dataframe, para melhor analisar seu conteúdo
pd.set_option('display.max_columns',74)
df.head(3)
```

```python
# Estabelecer apenas as colunas que quer trazer do dataset-dataframe
colunas = ['title' ,'upload_date'  ,'view_count'    ,'tags'      ,'uploader','description' ,'width'       ,'height'       ,'duration','tempo_desde_pub','query'] 
# Campos do scraping com BeautifulSoup4 e Requests :
# watch-title,watch-time-text,watch-view-count,og:video:tag,          ,og:description,og:video:width,og:video:height,          ,
```

```python
# Converter no formato binário com o feather :
# 1 facilita trabalhar com outras linguagens
# 2 a abertura do arquivo é mais rápida
df[colunas].to_feather('raw_data.to_feather')
```
